# DATA ANALYSIS


** What is the max Close price for each bank's stock throughout the time period?**

```python
bank_stocks.xs('Close', level=1, axis=1)          #.xs function()
```

```python
bank_stocks.xs('Close', level=1, axis=1).max()
```

** Creating a new empty DataFrame called returns, will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

** Can use pandas pct_change() method on the Close column to create a column representing this return value. **

```python
for x in tickers:
    returns[x + ' Return'] = bank_stocks.xs('Close',axis=1,level=1)[x].pct_change()
returns.head()
```

```python
import seaborn as sns
sns.set_style('whitegrid')     #to provide white background
```

```python
sns.pairplot(returns[1:])    #ploting above 5 columns
```

- In figure Citigroup's stock clearly stands out.
- This behavior was due their stock crashing during the financial crisis.
- the crash in value if you take a look at the stock price plot.

** Using this returns DataFrame,  what dates each bank stock had the best and worst single day returns. **

## Moving Averages

 moving averages for these stocks in the year 2008. 


```python
bank_stocks.loc['2008-01-01':'2008-12-31'].xs('Close',axis=1,level=1)['BAC'].plot(figsize=(12,6),label='BofA Close')
rolling_avg = pd.DataFrame()
rolling_avg['30 Day Avg'] = bank_stocks.loc['2008-01-01':'2008-12-31'].xs('Close',axis=1,level=1)['BAC'].rolling(window=30).mean()
rolling_avg['30 Day Avg'].plot(figsize=(12,6),label='30 Day Avg')
plt.legend()
```
